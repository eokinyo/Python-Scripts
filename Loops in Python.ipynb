## An infinite loop

```python
# Store the current largest number here.
largest_number = -999999999

# Input the first value.
number = int(input("Enter a number or type -1 to stop: "))

# If the number is not equal to -1, continue.
while number != -1:
    # Is number larger than largest_number?
    if number > largest_number:
        # Yes, update largest_number.
        largest_number = number
    # Input the next number.
    number = int(input("Enter a number or type -1 to stop: "))

# Print the largest number.
print("The largest number is:", largest_number)

```

```python
# A program that reads a sequence of numbers
# and counts how many numbers are even and how many are odd.
# The program terminates when zero is entered.

odd_numbers = 0
even_numbers = 0

# Read the first number.
number = int(input("Enter a number or type 0 to stop: "))

# 0 terminates execution.
while number != 0:
    # Check if the number is odd.
    if number % 2 == 1:
        # Increase the odd_numbers counter.
        odd_numbers += 1
    else:
        # Increase the even_numbers counter.
        even_numbers += 1
    # Read the next number.
    number = int(input("Enter a number or type 0 to stop: "))

# Print results.
print("Odd numbers count:", odd_numbers)
print("Even numbers count:", even_numbers)

```

```python
power = 1
for expo in range(16):
    print("2 to the power of", expo, "is", power)
    power *= 2

    '''
    This Java-like code demonstrates calculating powers of 2 up to 2^15.

1. `power = 1`: Initializes a variable `power` to the value 1.
2. `for expo in range(16):`: Initiates a loop that will execute 16 times, iterating through values from 0 to 15 (16 is excluded in Python).
3. `print("2 to the power of", expo, "is", power)`: Prints the current exponent value (`expo`), indicating the corresponding power of 2, along with the calculated value stored in `power`.
4. `power *= 2`: Updates the `power` variable by multiplying it by 2, effectively calculating the next power of 2 for the next iteration of the loop.
    '''
```

## The break and continue statemenst

```python
# break - example

print("The break instruction:")
for i in range(1, 6):
    if i == 3:
        break
    print("Inside the loop.", i)
print("Outside the loop.")


# continue - example

print("\nThe continue instruction:")
for i in range(1, 6):
    if i == 3:
        continue
    print("Inside the loop.", i)
print("Outside the loop.")

```

```python
largest_number = -99999999
counter = 0

while True:
    number = int(input("Enter a number or type -1 to end the program: "))
    if number == -1:
        break
    counter += 1
    if number > largest_number:
        largest_number = number

if counter != 0:
    print("The largest number is", largest_number)
else:
    print("You haven't entered any number.")

```

use conditional execution and the continue statement to "eat" the following vowels A, E, I, O, U from the inputted word;
assign the uneaten letters to the word_without_vowels variable and print the variable to the screen.
Look at the code in the editor. We've created word_without_vowels and assigned an empty string to it. Use concatenation operation to ask Python to combine selected letters into a longer string during subsequent loop turns, and assign it to the word_without_vowels variable.

Test your program with the data we've provided for you.


Test data:
Sample input:

```python
# Ask the user to enter a word
user_word = input("Enter a word: ")

# Convert the entered word to uppercase
user_word = user_word.upper()

# Initialize an empty string to hold uneaten letters
word_without_vowels = ""

# Loop through each character in the word
for letter in user_word:
    # Check if the letter is a vowel (A, E, I, O, U)
    if letter in ['A', 'E', 'I', 'O', 'U']:
        continue  # Skip the vowel and move to the next iteration
    
    # Concatenate uneaten consonants to the word_without_vowels variable
    word_without_vowels += letter

# Print the word containing uneaten consonants
print(word_without_vowels)
```

## Scenario
Listen to this story: a boy and his father, a computer programmer, are playing with wooden blocks. They are building a pyramid.

Their pyramid is a bit weird, as it is actually a pyramid-shaped wall â€“ it's flat. The pyramid is stacked according to one simple principle: each lower layer contains one block more than the layer above.

Your task is to write a program which reads the number of blocks the builders have, and outputs the height of the pyramid that can be built using these blocks.

Note: the height is measured by the number of fully completed layers â€“ if the builders don't have a sufficient number of blocks and cannot complete the next layer, they finish their work immediately.

Test your code using the data we've provided.


Test Data:
Sample input:

6
Expected output:

The height of the pyramid: 3

Sample input:

20
Expected output:

The height of the pyramid: 5

```python
blocks = int(input("Enter the number of blocks: "))  # Read the number of blocks as input and convert it to an integer

height = 0  # Initialize the height of the pyramid

# Loop to calculate the height of the pyramid
while blocks > 0:
    height += 1  # Increment the height for each completed layer
    blocks -= height  # Subtract the number of blocks used for the current layer
    
    if blocks < height + 1:  # Check if there are enough blocks to complete the next layer
        break  # If not enough blocks for the next layer, break the loop

print("The height of the pyramid:", height)  # Print the calculated height of the pyramid
```

## Scenario
In 1937, a German mathematician named Lothar Collatz formulated an intriguing hypothesis (it still remains unproven) which can be described in the following way:

take any non-negative and non-zero integer number and name it c0;
if it's even, evaluate a new c0 as c0 Ã· 2;
otherwise, if it's odd, evaluate a new c0 as 3 Ã— c0 + 1;
if c0 â‰  1, go back to point 2.
The hypothesis says that regardless of the initial value of c0, it will always go to 1.

Of course, it's an extremely complex task to use a computer in order to prove the hypothesis for any natural number (it may even require artificial intelligence), but you can use Python to check some individual numbers. Maybe you'll even find the one which would disprove the hypothesis.

Write a program which reads one natural number and executes the above steps as long as c0 remains different from 1. We also want you to count the steps needed to achieve the goal. Your code should output all the intermediate values of c0, too.

Hint: the most important part of the problem is how to transform Collatz's idea into a while loop â€“ this is the key to success.

```python
c0 = int(input("Enter a natural number: "))  # Read the initial number as input and convert it to an integer

steps = 0  # Initialize the steps counter

# Execute the Collatz conjecture steps until c0 becomes 1
while c0 != 1:
    print(c0)  # Print the current value of c0
    if c0 % 2 == 0:
        c0 //= 2  # If c0 is even, update c0 to c0 Ã· 2
    else:
        c0 = 3 * c0 + 1  # If c0 is odd, update c0 to 3 Ã— c0 + 1
    steps += 1  # Increment the steps counter

print(c0)  # Print the final value of c0 (which is 1)
print("steps =", steps)  # Print the total number of steps taken to reach 1
```
